```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from functools import reduce
```

```python
excel_sheets = pd.ExcelFile('KPMG_New_Raw_data.xlsx')
excel_sheets.sheet_names
trans_sheet = excel_sheets.parse(sheet_name='Transactions',header=1)
new_sheet = excel_sheets.parse(sheet_name='NewCustomerList',header=1,parse_dates=['DOB'])
demo_sheet = excel_sheets.parse(sheet_name='CustomerDemographic',header=1, parse_dates=['DOB'])
customer_sheet = excel_sheets.parse(sheet_name='CustomerAddress',header=1)
```

```python
sheets = [trans_sheet,demo_sheet,customer_sheet]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['customer_id'],
                                            how='inner'), sheets)
df_merged.head()
```

```python
df_merged.drop(columns='default', inplace=True)
df_merged.dropna(inplace=True) # df_merged.dropna(inplace=True)
```

```python
def pasrsing_datetime(df,columns):
    for col in columns:
        df[col]= pd.to_datetime(df[col])
    return df

columns = ['transaction_date','DOB']
pasrsing_datetime(df_merged,columns).head() 
```

```python
from datetime import datetime, date
def age(born):
    #born = datetime.strptime(born, "%Y/%m/%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
```

```python
# Drop null values
# Drop deceased is Yes
# gender column group it to one 
#

#create new columns
# 1 Age
# 2 Profit
# 3 recency column
# dropping default column as it contains meta data
bins= [19,29,39,49,59,69,79,89,99]
df_merged['age'] = df_merged['DOB'].apply(age)
df_merged['profit'] = df_merged['list_price']- df_merged['standard_cost']
df_merged['recency'] = pd.to_datetime('2017-12-30') - df_merged['transaction_date']
df_merged['grouped_age'] = pd.cut(df_merged['age'], bins, labels=['twenties','thirties','fourties','fifties','sixties','seventies','eighties','nineties'])
```

```python
demo_sheet['job_industry_category'].replace({'Argiculture':'Agriculture'}, inplace=True)
demo_sheet.drop('default', axis=1, inplace=True)
demo_sheet['age'] = demo_sheet['DOB'].apply(age)
demo_sheet['age_group'] =  pd.cut(demo_sheet['age'], bins, labels=['twenties','thirties','fourties','fifties','sixties','seventies','eighties','nineties'])
demo_sheet = demo_sheet[demo_sheet['deceased_indicator']=='N']
demo_sheet
```
